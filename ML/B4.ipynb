```python
def f(x):
  return (x + 3)**2

def df(x):
  return 2*(x + 3)
```

```python
def local_min(iterations, start, learning_rate, tolerance = 1e-6):
  x = start
  arr = []
  arr.append(x)

  for i in range(iterations):
    gradient = df(x)
    x_new = x - (learning_rate * gradient)
    arr.append(x_new)

    if abs(x_new - x) < tolerance :
      print(f"Convergence at iteration : {i + 1}")
      break

    x = x_new

  return x, arr
```

```python
start = 2
learning_rate = 0.1

local_minima, points = local_min(200, start, learning_rate)
```

```python
print(f"Local minima was found at x = {local_minima}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

x_vals = np.linspace(-10, 10, 100)

plt.plot(x_vals, f(x_vals))
plt.plot(points, f(np.array(points)), 'rx')

plt.xlabel('X')
plt.ylabel('f(X)')
plt.show()
```
